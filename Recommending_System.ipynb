```python
df_exploded  = grouped_data.explode('ItemId').drop('SessionId', axis=1)
```

```python
df_exploded.head()
```

```python
x_train = []
y_train = []
window_size = 5
for i in range(len(df_exploded) - window_size):
    x_train.append(df_exploded['ItemId'].iloc[i:i+window_size ].tolist())

    if i + window_size < len(df_exploded):
        y_train.append(df_exploded['ItemId'].iloc[i+window_size])
```

```python
print(len(x_train))
```

```python
unique_itemids = {}
encoded_y_train = []

# Loop through each itemid in y_train
for itemid in y_train:
  # Check if the itemid is already seen
  if itemid not in unique_itemids:
    # Assign a new encoded value (current length of the dictionary)
    encoded_value = len(unique_itemids)
    unique_itemids[itemid] = encoded_value
  else:
    # Use the existing encoded value for the itemid
    encoded_value = unique_itemids[itemid]
  
  # Append the encoded value to the encoded_y_train list
  encoded_y_train.append(encoded_value)

# Print the encoded data (unique itemids will be mapped to integers starting from 0)
print(encoded_y_train)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(
    x_train, encoded_y_train, test_size=0.1, random_state=3)
```

```python
y = tf.convert_to_tensor(Y_train,dtype=tf.int32)
x = tf.convert_to_tensor(X_train,dtype=tf.int32)
```

```python
y1=tf.convert_to_tensor(Y_test,dtype=tf.int32)
x1=tf.convert_to_tensor(X_test,dtype=tf.int32)
```

```python
#Simple-RNN
```

```python
model = keras.Sequential([
  layers.SimpleRNN(units=32, input_shape=(5,1)),
  layers.Dense(units=65044,activation='softmax')
])

# Compile the model
model.compile(loss='sparse_categorical_crossentropy', optimizer='adam',metrics=['accuracy', tf.keras.metrics.TopKCategoricalAccuracy(k=1)])
```

```python
model.summary()
```

```python
history=model.fit(x,y,epochs=10)
```
